###  Dictionary

Create Dictionary for following information as "person1":
   - Name = Srikanth
   - Gender= Male
   - Occupation = Data Scientist
   - Location  = Nizampet, Hyderabad

```python
# CODE HERE
person1 = {"Name": "Srikanth",
           "Gender": "Male",
           "Occupation": "Data Scientist",
           "Location": "Nizampet, Hyderabad"
           }
```

```python
print(person1)
```

> print keys in person1 as a list

```python
# CODE HERE
for key, value in person1.items():
    print(f"{key}")
```

```python
#  CODE HERE
lst = []
for key, value in person1.items():

    lst.append(key)
print(lst)
```

> Srikanth age is 27. Assign his age in person1 as key "Age"

```python
# CODE HERE
person1["Age"] = 27
```

```python
print(person1)
```

```python
person1
```

> Remove Location in person1

```python
# CODE HERE
person1.pop("Location")
```

```python
print(person1)
```
